#   Question 2 
Find max of open in each month


```python
columns1 = ["Date","open","close"]
data1 = [("2020-10-01 09:59:59", "100","105"), ("2020-10-03 09:59:59", "103","102"), ("2020-10-07 09:59:59", "104","110"), 
         ("2020-11-12 09:59:59", "110","112"), ("2020-11-16 09:59:59", "200","115"), ("2020-11-19 09:59:59", "95","92")]
dfFromData11 = spark.createDataFrame(data1).toDF(*columns1)
```

```python
import pyspark.sql.types as t
dfFromData12 = dfFromData11.withColumn("Date", col("Date").cast(t.TimestampType())).withColumn("open", col("open").cast(t.IntegerType()))
dfFromData12.show()
```

```python
from pyspark.sql.functions import month
df11 = dfFromData12.withColumn('month',month(dfFromData12.Date))
df11.show()
```

```python
df22 = df11.groupBy("month").agg(max(col("open")))
df22.show()
```
